## Matrices

### Price changes after deviation

* Agent 1

```python
%%time
pricechanges_a1_cut = pricechanges_dict(1,"cut",cycles,q_table_1,q_table_2,conv_info,S,A)
pricechanges_a1_raise = pricechanges_dict(1,"raise",cycles,q_table_1,q_table_2,conv_info,S,A)

# Concatenate dictionnaries
pricechanges_a1 = concatDict(pricechanges_a1_cut,pricechanges_a1_raise)
```

```python
# Processing for agent 1
## Retrieve preshock price
preshock_a1 = impulse_function(cycles,"cut",1,"full",q_table_1,q_table_2,conv_info,S,A,graph=False,n_seq=4,deviation_period=2)[0][:,1]

## Compute frequency and put in dataframe
freq = np.unique(preshock_a1, return_counts=True)
values = np.round(freq[0],2)
freq = freq[1]/sum(freq[1])
freq = pd.DataFrame(freq,index=values,columns=["Frequency"])

## Construct dataframe for relative price change matrix
mat = pd.DataFrame(pricechanges_mat(pricechanges_a1,A),index=np.round(A,2),columns=np.round(A,2))

## Concatenate previous dataframe
pricechanges_a1_df = pd.concat([freq, mat],axis=1)

## Replace NaN and remove empty rows
pricechanges_a1_df.replace(np.nan,0,inplace=True)
pricechanges_a1_df = pricechanges_a1_df.loc[(pricechanges_a1_df!=0).any(1)]
```
